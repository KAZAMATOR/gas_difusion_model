```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.animation import FuncAnimation
```

```python
data = np.loadtxt('datanew.txt')
```

```python
data.shape
```

```python
N = 6400
FRAMES = 100
```

```python
fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(projection='3d')
ax.set_xlim3d(0, 450)
ax.set_ylim3d(0, 450)
ax.set_zlim3d(0, 450)
ax.set_title('t = ')
ax.scatter([], [], [])
ax.scatter([], [], [])
```

```python
# ax.set_facecolor("#000066") 
```

```python
def animate_scatter(frame):
    data1 = data[frame*N:(frame+1)*N, :][data[frame*N:(frame+1)*N, 3]==1].T
    data0 = data[frame*N:(frame+1)*N, :][data[frame*N:(frame+1)*N, 3]==0].T #korzhik
    ax.cla()
    res1 = ax.scatter(data1[0], data1[1], data1[2], 'magenta')
    res0 = ax.scatter(data0[0], data0[1], data0[2], 'blue')
    ax.set_xlim3d(0, 250)
    ax.set_ylim3d(0, 250)
    ax.set_zlim3d(0, 150)
    ax.set_title(f't = {frame/50}')
    return res0, res1
```

```python
ax.cla()
anim = FuncAnimation(fig, animate_scatter, frames=FRAMES, interval=1, blit=False, repeat=True)
%time anim.save(f'out_2flags{N}.mp4', fps=30)
plt.show()
```
